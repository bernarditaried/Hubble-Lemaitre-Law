## Introduction

At the end of the 1920s, both Edwin Hubble and Georges Lemaitre discovered that galaxies farther away from us would move faster than nearby galaxies. They derived a linear relationship between the recession velocity $v$ and the distance from us $x$:

$$
v = H_0 x
$$

This is the famous <i>Hubble–Lemaitre law</i>. Its importance relies on the fact that it was the first observational evidence to support the Big Bang theory and an expanding universe. Some years before, in 1917, Albert Einstein proposed a static universe using General Relativity and introduced the so-called cosmological constant [1]. After realizing that observational data invalidate his model, Einstein himself called this his greatest mistake.

Nowadays, the estimated value for $H_0$ differs in a factor of 9 and 7 to the ones found by Lemaitre and Hubble, respectively. The inconsistencies are solved with a better calibration of the distances. But despite that, the linear model was correct, but it had different coefficients. This discovery changed our conception of the universe.

Hubble used a least square linear regression on 24 measurements of distances and recessional velocities of galaxies. In this tutorial, we will reproduce those results and introduce the least squares method.

Georges Lemaitre was a Belgian priest who published the first article about the universe's accelerated expansion. But he did it in a low-impact journal and in French [3] (I invite you to think about this and to read the following article: ["Does English Have to Be the Dominant Language of Science?"](https://blogs.scientificamerican.com/guest-blog/does-english-have-to-be-the-dominant-language-of-science/)). He observed a proportionality relation between the velocity of the galaxies moving off and their distance to us. He estimated this quantity and obtained its first estimate: $H_0$ = 625 km/s Mpc.

Two years later, in 1929, Edwin Hubble confirmed that the universe is expanding and determined the linear coefficient of the Hubble–Lemaitre law to be $H_0$ = 500 km/s Mpc. We nowadays call this parameter the Hubble constant $H_0$. 

Both cosmologists realized that the observations could be explained by the redshift effect derived in Friedmann's solutions to Einstein's equations of general relativity [5].

### Redshift

As light is an electromagnetic wave, it has an associated wavelength $\lambda$, and it travels through the space at a fixed velocity $c \sim 3 \cdot 10^{8} m/s$. The expansion of the universe elongates the light waves, and they end up with a different wavelength ($\lambda_{emit} < \lambda_{obs}$). Therefore, we will be observing a different type of light from the one emitted.

<img src="fig/redshift-and-blueshift.jpeg" width=700 height=700/> 

We call this effect <i>redshift</i>, because, in the electromagnetic spectra, in the optical range, the color red has a longer wavelength. The opposite of this effect is called <i>blueshift</i>, because blue as a shorter wavelength. We can measure this phenomena with the semi-positive quantity $z$ as follows:

$$
1 + z = \frac{\lambda_{obs}}{\lambda_{emit}}
$$

$$    
\lambda_{emit} + z \cdot \lambda_{emit} = \lambda_{obs}
$$

Therefore, if $z=0$, there would be no stretch at all and $\lambda_{emit} = \lambda_{obs}$. If $z=1$, it implies that $2 \lambda_{emit} = \lambda_{obs}$, which is equivalent to saying that the distances at the time of the observation are twice as they were original!

As objects tend to move apart (in reality, the space in between is expanding), we can define a quantity called the scale factor.

### Scale factor

The relative expansion of the universe is parametrized by a dimensionless time-dependent cosmic scale factor $a(t)$ as shown in the following figure:

<img src="fig/scale-factor-cat-and-galaxy.jpeg" width=500 height=500 />
<center>
We observe in this figure that the cat and the galaxy are moving further away in two moments: $t_1$ and $t_2 > t_1$. This is what happens in an expanding universe if bodies follow their so call Hubble flow. The comoving distance between points $x_1$ and $x_2$ remains constant, while the physical distance is proportional to the comoving distance times the scale factor $a(t)$. Figure adapted from [6]
</center>


The scale fator relates the proper distance $x(t)$ (which can change over time, unlike the comoving distance which is constant and set to today's distance) between two objects in different times. The formula for this is:

$$
x(t)=a(t)x(t_{0})
$$

where $x(t)$ is the proper distance at the time $t$, $x(t_{0})$ is the comoving distance ($t_{0}$ is set to be the present age of the universe), and $a(t)$ is the scale factor. By definition, $x(t_{0}) = x_{0}$ and $a(t_{0})=1$.


The evolution of the scale factor is more complicated than what we have exposed in here: it is determined by the equations of general relativity, which are related to the distribution of density of energy and other details. To quantify the change in the scale factor, we use the Hubble rate.

### Hubble rate:

The Hubble rate (or Hubble parameter) measures how "fast" the universe expands. It can be estimated by dividing the recession velocities of galaxies by their distance from us. This expansion rate is defined as:

$$
    H(t) = \frac{1}{a} \frac{da}{dt}
$$

where 

$$
    \frac{1}{H(t)} \propto [t]
$$

The Hubble rate varies with time, being the Hubble constant $H_{0}$ the current value:

$$
    H(t=0) = H_0
$$

We can measure the Hubble rate in units of velocity per distance:

$$
    H(t) \propto \left[ \frac{\text{km}}{\text{s} \cdot \text{Mpc}} \right]
$$

where Mpc = $10^6$ pc. The letters "pc" denote the parsec, defined as the distance at which one astronomical unit subtends an angle of one arcsecond. Also, 1 pc is equivalent to 3.26 light-years. 

If we calculate the derivative of $x(t)=a(t) x_{0}$, we obtain the velocity:

$$
    v(t) = \frac{d(x(t))}{dt} = \frac{a(t) x_{0}}{dt} = \frac{da}{dt} x_{0} 
$$

Where $\dfrac{d(x_{0})}{dt} = 0$ because it is constant. Then, if we substite the above definition of the Hubble rate gives:

$$
    v = \frac{da}{dt} x_0  = H_0 x(t) = H_0 x
$$

which is the famous Hubble–Lemaitre law. This relation is no longer valid for larger distances and velocities. It then depends on many other factors, such as the universe's energy content.

It is worth mentioning that what Lemaitre and Hubble measured were the apparent and absolute magnitudes ($m$ and $M$), not the distances directly. The apparent magnitude measures the brightness of an astronomical object observed from Earth. In contrast, the absolute magnitude is defined to be equal to the apparent magnitude that the object would have if it were viewed from a distance of 10 pc. They then derived the proper distance $x(t)$ between us and the galaxies:

$$
    x(t) = 10^{\left( \frac{m - M + 5}{5} \right)}
$$

However, the results for $H_0$ were higher becase Hubble used a wrong zero-point calibration of the standard candles. All distances were thus too small (by a factor of 7!), and then, $H_0$ was too large by the same factor.
