## Introduction

At the end of the 1920s, both Edwin Hubble and Georges Lemaître discovered that galaxies farther away from us would move faster than nearby galaxies. They derived a linear relationship between the recession velocity $v$ and the distance from us $x$:

$$
v = H_0 x
$$

This is the famous <i>Hubble–Lemaître law</i>. Its importance relies on the fact that it was the first observational evidence to support the Big Bang theory and an expanding universe. Some years before, in 1917, Albert Einstein proposed a static universe using General Relativity and introduced the so-called cosmological constant [1]. After realizing that observational data invalidate his model, Einstein himself called this his greatest mistake.

<div align="center"> The original plot made by Edwin Hubble in 1929 relates the distance and the recessive velocity of galaxies measured from the Earth [2]. </div>




Nowadays, the estimated value for $H_0$ differs in a factor of 9 and 7 to the ones found by Lemaître and Hubble, respectively. The inconsistencies are solved with a better calibration of the distances. But despite that, the linear model was correct, but it had different coefficients. This discovery changed our conception of the universe.

These are the questions I would have had in those days:
- How are these galaxies moving away from us?
- Are they doing it at a fixed velocity, or does it increase with the distance?
- How do we observe the light of galaxies that redshifted so much that they are not even visible in the optical range?

Hubble used a least square linear regression on 24 measurements of distances and recessional velocities of galaxies. In this tutorial, we will reproduce those results and introduce the least squares method.


## Cosmological context

### History

In the last century, we collected enough astronomical data to state that the universe is expanding. That means that, in the past, galaxies were closer. If we agree with the cosmological principle:

then galaxies should all be moving apart because there are no privileged points in the universe. This phenomenon was first studied by Lemaître [3] and Hubble [2]. They both looked at the shift of the light emitted by the stars from bright galaxies.

Edwin Hubble and Georges Lemaître

Georges Lemaître was a Belgian priest who published the first article about the universe's accelerated expansion. But he did it in a low-impact journal and in French [3] (I invite you to think about this and to read the following article: ["Does English Have to Be the Dominant Language of Science?"](https://blogs.scientificamerican.com/guest-blog/does-english-have-to-be-the-dominant-language-of-science/)). He observed a proportionality relation between the velocity of the galaxies moving off and their distance to us. He estimated this quantity and obtained its first estimate: $H_0$ = 625 km/s Mpc.

Two years later, in 1929, Edwin Hubble confirmed that the universe is expanding and determined the linear coefficient of the Hubble–Lemaître law to be $H_0$ = 500 km/s Mpc. We nowadays call this parameter the Hubble constant $H_0$. 

Both cosmologists realized that the observations could be explained by the redshift effect derived in Friedmann's solutions to Einstein's equations of general relativity [5].


### Redshift

As light is an electromagnetic wave, it has an associated wavelength $\lambda$, and it travels through the space at a fixed velocity $c \sim 3 \cdot 10^{8} m/s$. The expansion of the universe elongates the light waves, and they end up with a different wavelength ($\lambda_{emit} < \lambda_{obs}$). Therefore, we will be observing a different type of light from the one emitted.

<img src="fig/redshift-and-blueshift.png" width=700 height=700/> 

We call this effect <i>redshift</i>, because, in the electromagnetic spectra, in the optical range, the color red has a longer wavelength. The opposite of this effect is called <i>blueshift</i>, because blue as a shorter wavelength. We can measure this phenomena with the semi-positive quantity $z$ as follows:

$$
1 + z = \frac{\lambda_{obs}}{\lambda_{emit}}
$$

$$    
\lambda_{emit} + z \cdot \lambda_{emit} = \lambda_{obs}
$$

Therefore, if $z=0$, there would be no stretch at all and $\lambda_{emit} = \lambda_{obs}$. If $z=1$, it implies that $2 \lambda_{emit} = \lambda_{obs}$, which is equivalent to saying that the distances at the time of the observation are twice as they were original!

As objects tend to move apart (in reality, the space in between is expanding), we can define a quantity called the scale factor.

### Scale factor

The relative expansion of the universe is parametrized by a dimensionless time-dependent cosmic scale factor $a(t)$ as shown in the following figure:

<img src="fig/scale-factor-cat-and-galaxy.png" width=500 height=500 />
<center>
We observe in this figure that the cat and the galaxy are moving further away in two moments: $t_1$ and $t_2 > t_1$. This is what happens in an expanding universe if bodies follow their so call Hubble flow. The comoving distance between points $x_1$ and $x_2$ remains constant, while the physical distance is proportional to the comoving distance times the scale factor $a(t)$. Figure adapted from [6]
</center>


The scale fator relates the proper distance $x(t)$ (which can change over time, unlike the comoving distance which is constant and set to today's distance) between two objects in different times. The formula for this is:

$$
x(t)=a(t)x(t_{0})
$$

where $x(t)$ is the proper distance at the time $t$, $x(t_{0})$ is the comoving distance ($t_{0}$ is set to be the present age of the universe), and $a(t)$ is the scale factor. By definition, $x(t_{0}) = x_{0}$ and $a(t_{0})=1$.


The evolution of the scale factor is more complicated than what we have exposed in here: it is determined by the equations of general relativity, which are related to the distribution of density of energy and other details. To quantify the change in the scale factor, we use the Hubble rate.

### Hubble rate:

The Hubble rate (or Hubble parameter) measures how "fast" the universe expands. It can be estimated by dividing the recession velocities of galaxies by their distance from us. This expansion rate is defined as:

$$
H(t) = \frac{1}{a} \frac{da}{dt}
$$

where 

$$
\frac{1}{H(t)} \propto [t]
$$

The Hubble rate varies with time, being the Hubble constant $H_{0}$ the current value:

$$
H(t=0) = H_0
$$

We can measure the Hubble rate in units of velocity per distance:

$$
H(t) \propto \left[ \frac{\text{km}}{\text{s} \cdot \text{Mpc}} \right]
$$

where Mpc = $10^6$ pc. The letters "pc" denote the parsec, defined as the distance at which one astronomical unit subtends an angle of one arcsecond. Also, 1 pc is equivalent to 3.26 light-years. 

If we calculate the derivative of $x(t)=a(t) x_{0}$, we obtain the velocity:

$$
v(t) = \frac{d(x(t))}{dt} = \frac{a(t) x_{0}}{dt} = \frac{da}{dt} x_{0} 
$$

Where $\dfrac{d(x_{0})}{dt} = 0$ because it is constant. Then, if we substite the above definition of the Hubble rate gives:

$$
v = \frac{da}{dt} x_0  = H_0 x(t) = H_0 x
$$

which is the famous Hubble–Lemaître law. This relation is no longer valid for larger distances and velocities. It then depends on many other factors, such as the universe's energy content.

It is worth mentioning that what Lemaître and Hubble measured were the apparent and absolute magnitudes ($m$ and $M$), not the distances directly. The apparent magnitude measures the brightness of an astronomical object observed from Earth. In contrast, the absolute magnitude is defined to be equal to the apparent magnitude that the object would have if it were viewed from a distance of 10 pc. They then derived the proper distance $x(t)$ between us and the galaxies:

$$
x(t) = 10^{\left( \frac{m - M + 5}{5} \right)}
$$

However, the results for $H_0$ were higher becase Hubble used a wrong zero-point calibration of the standard candles. All distances were thus too small (by a factor of 7!), and then, $H_0$ was too large by the same factor.

### Method of least squares

The <i>method of least squares</i> was developed independently by Gauss, Legendre, and Adrain [6]. This method allows us to find a regression equation that relates a dependent variable with one or more independent variables. The regression can be linear or non-linear, and the optimization of the parameters is obtained by deriving the minimum of the sum of squares of the residuals (i.e. deviations from the regression).

Let us have $n$ observations: $X_1, X_2, ..., X_n$, such that the number of parameters to estimate $p$ is smaller than the number of observations ($p<n$). Each observation will have a random error (i.e. following a Gaussian distribution) associated as:

$$
    X_i = \alpha + \epsilon_i
$$

where $X_i$ would be the i-th measurement, $\alpha$ the true value of a physical quantity, and $\epsilon_i$ the error in the i-th measurement. 

Our estimators for both $\alpha$ and $\epsilon_i$ are:

$$
\hat{\alpha} = a \longrightarrow \text{ estimate of } \alpha
$$

$$
\hat{\epsilon_i} = e_i \longrightarrow \text{ estimate of } \epsilon_i
$$

This would be equivalent to have:

$$
    X_i = \alpha + \epsilon_i = a + e_i
$$

The method's goal is to estimate the regression coefficients (in this case $\alpha$ and $\epsilon_i$) by minimizing the sum of the squares (or other quadratic forms). The sum of the squares is

$$
    \sum_{i=1}^n e_i^2 = \sum_{i=1}^n (a-X_i)^2
$$

The minimum is obtained as

$$
    \frac{\partial(\sum_{i=1}^n e_i^2)}{\partial a} = 0 \Rightarrow a = \sum_{i=1}^n \frac{y_i}{n} = \overline{y}
$$

which the last term is equal to the arithmetic mean of the observations $\overline{y}$, returning a simple estimate of $\alpha$. The corresponding estimate of the dispersion is the standard deviation from $\overline{y}$. The simplest case is when we apply a linear regression to a set of points ($x_i, y_i$) such that they follow a linear relation:

$$
    y = \alpha + \beta x
$$

and the values $x_i$ are exact, while the values $y_i$ are subjected to some error $\epsilon_i$. Following the same approach, the observational equations would be given as:

$$
    y_i =  \alpha + \beta x_i + \epsilon_i = a + b x_i + e_i
$$

where $a$, $b$, and $e_i$ are estimates of $\alpha$, $\beta$ and $\epsilon_i$ respectively. The optimization should be done on:

$$
    \sum_{i=1}^n e_i^2 = \sum_{i=1}^n (a + b x_i - y_i)^2
$$

We minimize both $a$ and $b$ by optimizing:

$$
\frac{\partial(\sum_{i=1}^n e_i^2)}{\partial a} = 0 \Rightarrow n a + b \sum_{i=1}^n x_i = \sum_{i=1}^n y_i
$$

$$
\frac{\partial(\sum_{i=1}^n e_i^2)}{\partial b} = 0 \Rightarrow a \sum_{i=1}^n x_i + b \sum_{i=1}^n x_i^2 = \sum_{i=1}^n x_i y_i
$$

The solution would be:


$$
    a = \frac{n \sum_{i=1}^n x_i y_i - \sum_{i=1}^n x_i \sum_{i=1}^n y_i}{n \sum_{i=1}^n x_i^2 - ( \sum_{i=1}^n x_i)^2}
$$

$$
b = \frac{\sum_{i=1}^n x_i^2 \sum_{i=1}^n y_i - \sum_{i=1}^n x_i \sum_{i=1}^n x_i y_i}{n \sum_{i=1}^n x_i^2 - ( \sum_{i=1}^n x_i )^2}
$$

were $a$ and $b$ would be the $y$-intersection and the slope of the linear regression respectively. 


We finally proceed to compute the linear regression of the data on Hubble's observations.

```python
n = len(df) # Number of observatios

s_x = np.sum(x) # Sum over x
s_y = np.sum(y) # Sum over y

s_xx = x.dot(x) # Sum over x^2
s_xy = x.dot(y) # Sum over xy

a = (s_xx * s_y - s_x * s_xy)/(n * s_xx - (s_x)**2) # a coefficient
b = (n * s_xy - s_x * s_y)/(n * s_xx - (s_x)**2) # b coefficient

x_plot = np.linspace(0, 2, 20)
y_plot = a+x_plot*b

H_0 = round(b, 1)

fig, ax = plt.subplots(figsize=(8, 5))
plt.plot(x, y, 'd', markersize=10, color=color_3, label='Data')
plt.plot(x_plot, y_plot, color=color_2, label = 'Least squares fit')
plt.xlabel('Distance [Mpc]')
plt.ylabel('Velocity' +  r' [km s$^{-1}$]')
plt.title('Hubble–Lemaître Law')
ax.text(1.24, -215, r"$H_0 = $ " + str(H_0) + r' [km s$^{-1}$ Mpc$^{-1}$]', size='x-large', color=color_2)
plt.legend()
plt.show()
```

We obtain a value of $H_0 = 454.2$ [km s$^{-1}$ Mpc$^{-1}$] with Hubble's data. The age of the universe would be simply the inverse of the Hubble constant:

$$    
t_0 = \frac{1}{H_0}
$$

```python
# Convert H_0 from Mpc to km:
H_0_km = H_0 / 1000000 / 30856775814913.67
t_0_sec = (1/H_0_km)

# Age of the universe in years
t_0 = t_0_sec / (60*60*24*365)

print('The age of the universe obtained is: ' + "{:,}".format(int(t_0)) + ' years.')
```

Our latest universe age estimations are near the 13.8 billion years in the $\Lambda$CDM model. This model is based on further details, such as the energy densities present through the history of the cosmos. The Hubble–Lemaître law works locally; corrections must be made when observing deeper astronomical objects. We must consider that the observations are also subject to systematic errors.
